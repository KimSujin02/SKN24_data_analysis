### merge

특정 컬럼 기준으로 DataFrame 병합
```
 pd.merege(df1, df2, on, how)
```


- on : merge 기준 컬럼(지정하지 않으면 같은 이름의 컬럼을 기준으로 함)
- how : inner(기본값), outer, left, right


````python
df1 = pd.DataFrame({
    'id' : [1,2,3,4],
    'name' : ['다람쥐','토기','개구리','카피바라'],
})

df2 = pd.DataFrame({
    'id' : [1,2,3,4],
    'class_no' : [4,2,1,7]
})
````

````python
pd.merge(df1,df2)
````

### join
- 특정 인덱스 기준으로 병합을 한다
```
df1.join(df2)
```
- how : inner, left(기본값), right, outer

````python
df3 = pd.DataFrame({
    'voice_key' : [1,3,5,7],
    'city' :['서울','수원','인천','서울']
}, index=['박효신','김범수','나얼','남윤진'])
df4 = pd.DataFrame({
    'pet' : ['고양이','강아지','햄스터','다람쥐']
}, index = ['박효신','김범수','나얼','윤진'])
````

````python
df4
````

````python
df3.join(df4)
````

````python
df3.join(df4,how='left')
````

````python
df3.join(df4,how='right')  # 오른쪽 기준
````

````python
df3.join(df4,how='outer')
````

````python
df3.join(df4,how='inner')
````

### concat
- 축을 기준으로 병합

````python
df5 = pd.DataFrame({
    'name' : ['구준표','윤지후','소이정','프린스송'],
    'age' : [18,18,18,18]
})
df6 = pd.DataFrame({
    'name' : ['금잔디'],
    'age' : [17]
})
````

````python
pd.concat([df5,df6], axis=0).reset_index(drop=True)
````

````python
a =pd.concat([df5,df6], axis=1).reset_index(drop=True)
a
````

### Melt
- 열 병합

````python
df = pd.DataFrame({
    'name' : ['RM', '정국', '지민', '뷔'],
    'math' : [80, 90, 100, 100],
    'english' : [100, 90, 80, 60],
    'science' : [99, 77, 43, 100]
})
````

````python
pd.melt(
    df,
    id_vars = ['name'],                         # 고정 (유지) 될 컬럼
    value_vars= ['math','english', 'science']   # 합쳐질 컬럼
)
````

````python
pd.melt(
    df,
    id_vars = ['name'],                         # 고정 (유지) 될 컬럼
    value_vars= ['math','english', 'science'],  # 합쳐질 컬럼
    var_name = 'subject',
    value_name= 'score'
)
df
````
