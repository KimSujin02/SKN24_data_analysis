# 통계 관련 NumPy 활용

```python
import numpy as np
```

### 중심 경향 측도

```python
height = np.array([160, 170, 180, 190, 150, 173, 172, 182, 184, 155, 165, 155, 165, 175, 185, 195])
```

```python
count = height.size
height_sum = height.sum()
print(count, height_sum, height_sum/count)

# 평균값
print(np.mean(height))
```

```python
print(np.sort(height))

# 중위값 (중앙값)
print(np.median(height))
```

```python
# 최빈값 (unique() + max() + Boolean Indexing)
values, counts = np.unique(height, return_counts=True) # 중복값 배제
# return_counts=True 옵션을 주면
print(values)
print(counts)

# 값이 제일 많은 배열의 요소를 뽑기
max_count = np.max(counts)
print(counts == max_count)
print(values[counts == max_count])

# 이렇게 모듈을 활용해서도 가능!
from scipy import stats

print(stats.mode(height))
```

### 산포 측도

```python
# 분산 variance
# (값 - 평균)^2의 평균
var1 = np.mean(np.square(height - np.mean(height)))
print(var1)

var2 = np.var(height)
print(var2)
```

```python
# 표준편차 standard deviation
# 분산의 제곱근
std1 = np.sqrt(var1)
print(std1)

std2 = np.std(height)
print(std2)
```

```python
# 범위 (range)
# 값의 규모
range = np.max(height) - np.min(height)
print(range)
```

### 확률분포

```python
# !pip install matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
# 정규분포
# np.random.randn(shape) : (평균이 0, 표준편차가 1인) 표준 정규분포를 따르는 난수 생성
std_arr = np.random.randn(1000)
plt.hist(std_arr, bins=100)
plt.show()
```

```python
# 정규분포
# np.random.normal(loc, scale, size)
# - loc : 평균
# - scale : 표준편차
# - size : shape(갯수)
npr_arr = np.random.normal(10, 1, 1000)
plt.hist(npr_arr, bins=100)
plt.show()
```

```python
# 균등분포
# np.random.rand(num) : 0.0이상 1.0 미만 수에서 균등분포를 따르도록 난수 생성
# - num : shape(갯수)
arr = np.random.rand(1000)
plt.hist(arr, bins=100)
plt.show()
```

```python
# 균등분포
# np.random.randint(start, end, size)
# start 이상 end 미만의 균등분포를 따르는 난수생성
# - start : 평균
# - end : 표준편차
# - size : shape(갯수)

np.random.seed(0)       # 난수 계산 : 시작값 seed 고정

arr = np.random.randint(1, 100, 1000)
plt.hist(arr, bins=10)
plt.show()
```
