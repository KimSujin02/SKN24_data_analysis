### **[실습] 기본 실습**

| Name | Age | City | Salary |
| --- | --- | --- | --- |
| Alice | 25 | New York | 70000 |
| Bob | 30 | Los Angeles | 80000 |
| Charlie | 35 | Chicago | 120000 |
| David | 40 | Houston | 90000 |
| Eve | 28 | San Francisco | 75000 |
- 다음 데이터를 활용하여 실습 문제를 해결하세요.

1. **DataFrame객체 생성 :** 위의 데이터를 사용해 DataFrame객체를 생성하세요.

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    'Name':['Alice', 'Bob', 'Charlie', 'David', 'Eve']
    , 'Age':[25, 30, 35, 40, 28]
    , 'City':['New York', 'Los Angeles', 'Chicago', 'Houston', 'San Francisco']
    , 'Salary':[70000, 80000, 120000, 90000, 75000]
})
df
```

2. **열과 행 선택**
    - `Name`과 `City` 열을 선택하여 출력하세요.
    - 나이가 30 이상인 행을 선택하여 출력하세요.

```python
print(df[['Name', 'City']])
print(df[df['Age'] > 30])
```

3. **행과 열 추가**
    - `Country` 열을 추가하고, 모두 'USA'로 설정하세요.
    - 새로운 행을 추가하세요. (Name: Frank, Age: 33, City: Seattle, Salary: 85000)

```python
df['Country'] = 'USA'
df.loc[len(df)] = ['Frank', 33, 'Seattle', 85000, 'USA']
df
```

4. **데이터 삭제**
    - `Salary` 열을 삭제하세요.
    - 'David'가 있는 행을 삭제하세요.

```python
# df.drop('Salary', axis=1, inplace=True)
df.drop(df.index[df['Name'] == 'David'], inplace=True)
df
```

5. **데이터 수정**
    - `Age`가 35 이상인 사람들의 `Salary`를 5% 증가시키세요.
    - 'Alice'의 `City`를 'Boston'으로 수정하세요.

```python
df['Salary'] = np.where(df['Age'] >= 35, (df['Salary']*(1+(5/100))), df['Salary'])
df.loc[df.index[df['Name'] == 'Alice'], 'City'] = 'Boston'
df
```

6. **데이터 타입 변환**
    - `Age`와 `Salary`의 데이터 타입을 float로 변환하세요.

```python
df['Age'] = df['Age'].astype(float)
df['Salary'] = df['Salary'].astype(float)
df
```
