### 중복 데이터 처리

```python
df.duplicated().sum()  
# 중복 == 모든 컬럼 값이 같을 때 True를 반환을 하게 됨-> row 자체가 복제됐다고 생각하면 됨
#df.duplicated().sum() : True인 값을 세려줌 .중복 데이터의 수 
# 결과가 1 == 중복되는 데이터가 1개 == 같은 데이터가 2개
# - 결과가 2 == 중복 되터가 2개 == A라는 데이터가 3개 있는 경우 or A 라는 데이터가 2개, B 데이터가 2개 있는 경우



```

```python
#df.duplicated() : 기본적으로 모든 컬럼값이 같을 때 중복으로 간주해 중복 row를 True로 하는 Bool Series 반환
# 특정 컬럼을 지정해 준복 뎅치터 찾기
df[df.duplicated('Name')]
```

```python
df = df.drop_duplicates()
df
```

```python
# 중복 데이터 제거
df = df.drop_duplicates().reset_index(drop=True)
                            # rest_index() : 제거된 행을 제외하고 인덱스 재설정
                            #               기존 인덱스를 컬럼으로 추가하지 않으려면
                            #               drop =True 인자를 전달해야 함
df
```

### 결측치 처리

```python
df.isna().sum() # boolean 데이터 프레
# df.isnull()
```

```python
# 결측치 처리 1 : 대표값으로 결측치 대체 (평균값, 기본값, 최빈값,...)
df['Phone'] = df['Phone'].fillna('010-0000-0000') # 기본값 대체

# df['Phone'].fillna('010-0000-0000') 이거 자체가 series 형태이기 때문에 
#df =df['Phone'].fillna('010-0000-0000') 로 하면 df는 더이상 dataframe이 아니라 sereis가 됨 그래서 
# df['Phone'] = df['Phone'].fillna('010-0000-0000') 으로 해줘야 됨

df['Phone'].value_counts
df.isnull().sum()

print(df['Phone'].value_counts())
```

```python
# 결측값 처리 2 : 결측치가 포함된 행 제거
# - how='all' : 모든 셀이 결측치인 행 제거
# = how='any' : 결측치 셸을 하나라도 가지고 있으면 행 제거 (기본값)
df = df.dropna(how='all')
df.info()
```
